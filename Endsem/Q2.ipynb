```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the newly uploaded dataset
file_path = './SOUTHBANK.csv'
southbank_data = pd.read_csv(file_path)

# Step 2: Define the first 100 rows as training data
training_data = southbank_data.iloc[:100]

# Extract features (X) and target (y)
X_train = training_data[['OPEN']]  # Feature: OPEN price
y_train = training_data['LOW']     # Target: LOW price

# Step 3: Train a Simple Linear Regression (SLR) model
slr_model = LinearRegression()
slr_model.fit(X_train, y_train)

# Step 4: Extract Regression Coefficients
beta_0 = slr_model.intercept_  # Intercept
beta_1 = slr_model.coef_[0]    # Slope (Coefficient for OPEN price)

# Step 5: Predict the LOW price for the 101st day
test_data = southbank_data.iloc[100:101]  # 101st day data
X_test = test_data[['OPEN']]              # OPEN price for the 101st day
predicted_low_price = slr_model.predict(X_test)[0]

# Step 6: Delta change in predicted LOW price for every unit change in OPEN price
delta_change = beta_1

# Step 7: Output Results
print("Regression Coefficients:")
print(f"  Beta_0 (Intercept): {beta_0}")
print(f"  Beta_1 (Slope): {beta_1}")
print("\nPredicted LOW price for the 101st day:")
print(f"  {predicted_low_price}")
print("\nDelta change in LOW price for every unit change in OPEN price:")
print(f"  {delta_change}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

if not os.path.exists('./Images'):
    os.makedirs('./Images')

# Generate predictions for the training data
y_train_pred = slr_model.predict(X_train)

# Scatter plot of actual vs predicted values for training data
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Actual LOW Prices (Training Data)')
plt.plot(X_train, y_train_pred, color='red', label='Regression Line (SLR Model)')
plt.title("Simple Linear Regression: Predicted vs Actual (Training Data)")
plt.xlabel("OPEN Price")
plt.ylabel("LOW Price")
plt.legend()
plt.grid()
plt.savefig('./Images/slr_training_data.png')
plt.show()
```
